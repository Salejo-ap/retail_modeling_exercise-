# Limpiea de datos y transforamcion de datos

```python
# se filtran los datos los cuales no tengan reward y que no tengan tiendas
df_clean_1= df[df['Reward'] != 0]
df_clean_1= df_clean_1[df['NumStores'] != 0]
df_clean_1.describe()
```

```python
# elimar las columnas donde el mes de expiracion sea 999 
df_clean_2 = df_clean_1[df_clean_1['ExpirationMonth'] != 999]
df_clean_2.describe()
```

```python
variables_categoricas = ['Industry','ProgramName','RewardType','RewardStructure','RewardSize','ExpirationMonth','IndustryType']
for var in variables_categoricas:
    print(f'variable: {var}')
    print(df_clean_2[var].unique())
```

```python
# Convertir la columna Expiration y RewardSize a tipo de dato int y float respectivamente
df_clean_2['ExpirationMonth'] = df_clean_2['ExpirationMonth'].astype(int)
df_clean_2['RewardSize'] = df_clean_2['RewardSize'].astype(float)
df_clean_2.describe()
```

```python
# dataset final para el modelado
df_cleaned = df_clean_2[[ 'Salerank', 'X2013USSales', 'X2013WorldSales', 'ProfitMargin', 'NumStores', 'Reward', 'RewardSize', 'ExpirationMonth' ]]
```

# Exploracion de los datos

# Modelado regresion lineal simple

```python
# importacion modelo de regresion lineal 
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```python
df_cleaned.info()
```

```python
# Desplegar el modelo identificando la  mejor variable predictora para la variable objetivo ExpirationMonth
predictores =['Salerank', 'X2013USSales', 'X2013WorldSales', 'ProfitMargin', 'NumStores', 'Reward', 'RewardSize']
y = df_cleaned['ExpirationMonth']
for col in predictores:
    X = df_cleaned[[col]]
    Lm= LinearRegression()
    Lm.fit(X, y)
    y_pred = Lm.predict(X)
    r2 = r2_score(y, y_pred)
    print(f'predictores: {col}, R2: {r2}')
```

### Modelo con la variable predictora que escogimos 

```python
# Modelado y transformacion de la variable predictora NumStores
x=df_cleaned[['NumStores']]
y=df_cleaned['ExpirationMonth']
constructor=LinearRegression()
constructor.fit(x,y)
y_pred=constructor.predict(x)
r2=r2_score(y,y_pred)
intercepto=constructor.intercept_
coeficiente=constructor.coef_
print(f'R2: {r2}, Intercepto: {intercepto}, Coeficiente: {coeficiente}')
```

### modelo con la misma variable, aplicando logaritmo a la variable objetivo

```python
y = np.log(df_cleaned['ExpirationMonth'])
constructor=LinearRegression()
constructor.fit(x,y)
y_pred=constructor.predict(x)
r2=r2_score(y,y_pred)
intercepto=constructor.intercept_
coeficiente=constructor.coef_
print(f'R2: {r2}, Intercepto: {intercepto}, Coeficiente: {coeficiente}')
```

### modelo con la misma variable, aplicando logaritmo a la variable predictora

```python
x=np.log(df_cleaned[['NumStores']])
y = df_cleaned['ExpirationMonth']
constructor=LinearRegression()
constructor.fit(x,y)
y_pred=constructor.predict(x)
r2=r2_score(y,y_pred)
intercepto=constructor.intercept_
coeficiente=constructor.coef_
print(f'R2: {r2}, Intercepto: {intercepto}, Coeficiente: {coeficiente}')
```

### modelo con la misma variable, aplicando logaritmo a las dos variables


```python
x=np.log(df_cleaned[['NumStores']])
y = np.log(df_cleaned['ExpirationMonth'])
constructor=LinearRegression()
constructor.fit(x,y)
y_pred=constructor.predict(x)
r2=r2_score(y,y_pred)
intercepto=constructor.intercept_
coeficiente=constructor.coef_
print(f'R2: {r2}, Intercepto: {intercepto}, Coeficiente: {coeficiente}')
```
